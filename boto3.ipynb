```python
import boto3

s3 = boto3.client(
    "s3",
    endpoint_url="http://localhost:9000",
    aws_access_key_id="admin",
    aws_secret_access_key="admin123",
)

s3.upload_file(
    Filename="C:\\Users\\A923apulia\\Desktop\\Megabonk.url",
    Bucket="bucket",
    Key="Megabonk.url"
)

print("Upload Completo")
```

```python
s3.download_file(
    Bucket="bucket",
    Key="Megabonk.url",
    Filename="Megabonk.url"
)

print("Download Completo")
```
